# Crossword Puzzle

A **crossword puzzle** is a word game in which the goal is to fill a grid made of white and black squares with letters so that they form words that match given clues. The puzzle provides clues for words placed across and down, and the solver must write answers in the white squares with lengths that match the available spaces. As words intersect, shared letters must be consistent for both words. The puzzle begins with an empty or partially filled grid and is solved when all white squares are correctly filled, every word fits its clue, and all intersecting letters match, producing a fully consistent grid that satisfies all the puzzle’s constraints.

## Instance description

The instance of crossword puzzle consists of a *n*×*n* grid and the candidate words that must be placed into the grid. The grid contains two types of cells: blank cells (_), where letters must be filled, and blocked cells (#), which separate words and cannot contain letters. Words are formed by sequences of blank cells arranged horizontally (across) or vertically (down) between blocked cells or grid boundaries. Each sequence of blanks defines a slot whose length determines which words from the provided list can fit in that position.
